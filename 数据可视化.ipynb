# Python数据可视化

## 单变量可视化

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 示例数据
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [200, 220, 250, 270, 300, 320]
}
df = pd.DataFrame(data)

# 使用Matplotlib绘制折线图
plt.figure(figsize=(5, 3))
plt.plot(df['Month'], df['Sales'], marker='o') # marker='o'表示使用圆点标记
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales') 
plt.grid(True) # 显示网格
plt.show()

# 使用seaborn绘制直方图
# 生成随机数据
data = np.random.randn(1000)

# 绘制直方图
plt.figure(figsize=(5, 3))
sns.histplot(data, bins=30, kde=True, color='blue')
plt.title('Histogram of Random Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# 使用Seaborn绘制条形图
plt.figure(figsize=(5, 3))
sns.barplot(x='Month', y='Sales', data=df)
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

# 使用Seaborn绘制折线图
plt.figure(figsize=(5, 3))
sns.lineplot(x='Month', y='Sales', data=df, marker='o') # marker='o'表示使用圆点标记
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True) # 显示网格
plt.show()


```

## 两变量的关系
* 连续与连续：散点图
* 连续与离散
    * y分类，x连续：机器学习分类模型
    * y连续，x离散：箱线图boxplot
* 离散与离散：列联表检验如卡方检验，可以画马赛克图
* 多变量的关系：相关系数矩阵热力图

```python
# 使用Seaborn绘制散点图
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 假设的数据
data = {
    'House Size (sq ft)': [1500, 1800, 2400, 3000, 3500],
    'Price (000s)': [400, 500, 600, 650, 700]
}
df = pd.DataFrame(data)

# 绘制散点图
plt.figure(figsize=(5, 3))
sns.scatterplot(x='House Size (sq ft)', y='Price (000s)', data=df)
plt.title('Relationship Between House Size and Price')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (in 000s)')
plt.show()
```

```python
# 使用Seaborn绘制箱线图
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 假设数据
data = {
    'Group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Score': [85, 90, 78, 88, 77, 92, 84, 76, 95]
}
df = pd.DataFrame(data)

# 绘制箱线图
plt.figure(figsize=(5, 3))
sns.boxplot(x='Group', y='Score', data=df)
plt.title('Exam Scores by Group')
plt.xlabel('Group')
plt.ylabel('Score')
plt.show()
```

```python
# 使用Seaborn绘制小提琴图
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 假设的数据
data = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Weight': [68, 55, 75, 60, 82, 65, 77, 58]
}
df = pd.DataFrame(data)

# 绘制小提琴图
plt.figure(figsize=(5, 3))
sns.violinplot(x='Gender', y='Weight', data=df)
plt.title('Weight Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# 创建数据
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Smoking': ['Yes', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)

# 创建列联表
contingency_table = pd.crosstab(df['Gender'], df['Smoking'])

# 执行卡方检验
chi2, p, dof, expected = chi2_contingency(contingency_table)

# 打印卡方检验结果
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# 绘制马赛克图
plt.figure(figsize=(5, 3))
props = lambda key: {'color': 'red' if 'Yes' in key else 'blue'}  # 可以根据需要调整颜色
mosaic(df, ['Gender', 'Smoking'], gap=0.02, properties=props, title='Mosaic Plot of Gender vs Smoking')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 假设的数据
data = {
    'Age': [25, 30, 35, 40, 45],
    'Height': [175, 180, 165, 170, 175],
    'Weight': [70, 80, 60, 65, 75],
    'Income': [50000, 60000, 55000, 65000, 70000]
}

df = pd.DataFrame(data)

# 计算相关系数矩阵
correlation_matrix = df.corr()

# 创建热力图
plt.figure(figsize=(5, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
